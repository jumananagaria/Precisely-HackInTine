```python
#importing the libraries
import pandas as pd
```

```python
#reading the dataset
data=pd.read_csv("Opportunity 2.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
#Converting the whole dataset into smaller case

data = data.apply(lambda x: x.astype(str).str.lower())
```

```python
data.head()
```

```python
data.dtypes
```

```python
data = data.dropna(how='any')
data
```

```python
#creating another column with discipline and subtags
comb_frame = data['Discipline'].str.cat(" "+data['Sub tags'])
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#removing stop words
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
X = vectorizer.fit_transform(comb_frame)
```

```python
# true_k, derived from elbow method
true_k = 3
```

```python
#building a model
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=500, n_init=15)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :15]:
        print(' %s' % terms[ind]),
    print
```

```python
import matplotlib.pyplot as plt
# Continuing after vectorization step
# data-structure to store Sum-Of-Square-Errors
sse = {}
# Looping over multiple values of k from 1 to 30
for k in range(1, 40):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=100).fit(X)
    comb_frame["clusters"] = kmeans.labels_
    sse[k] = kmeans.inertia_
# Plotting the curve with 'k'-value vs SSE
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
# Save the Plot in current directory
plt.savefig('elbow_method.png')
```
